#### Evaluation Metrics

We have three major types of Machine Learning Algorithms

- Regression (Supervised)
- Classification (Supervised)
- Clustering (Unsupervised)

All of which have different evaluation metrics

### Regression

- Mean Absolute Error:
    - MAE measures the average magnitude of the error
    - Actual Data Point - Predicted Data Point of all points
    - MAE = (1/n) * Σ |yi − ŷi|
    - Lower the MAE, the better the model

- Mean Squared Error:
    - MSE measures the average of the squares of the errors
    - MSE = (1/n) * Σ (yi − ŷi)^2
    - again Lower the MSE, the better the model
    - More sensitive to outliers than MAE

- Root Mean Squared Error:
    - RMSE is the square root of MSE. 
    - A way to quantify the size of the error made by the model
    - RMSE = √((1/n) * Σ (yi − ŷi)^2)
    - More sensitive to outliers than MAE
    - again Lower the RMSE, the better the model

- R-squared (Coefficient of determination):
    - R^2 represents the proportion of the variance or the dependent variable that's explained by the independent variables.
    - R² = 1 − [ Σ (yi − ŷi)^2 / Σ (yi − ȳ)^2 ]
    - Values range from 0 to 1.
    - A higher R-squared indicates a better fit/model

- Adjusted R-squared:
    - Modified version o R-squared that has been adjusted for the number of predictors in the model.
    - Accounts for the model complexity
    - Adjusted R² = 1 − [ (1 − R²)(n − 1) / (n − p − 1) ]
    - If a new feature:
        - Actually improves the model → Adjusted R² increases
        - Adds noise → Adjusted R² decreases

- Mean Absolute Percentage Error
    - MAPE provides error in terms of percentage, making it scale-independent
    - MAPE = (100/n) * Σ |(yi − ŷi) / yi|

#### Choosing the Right regression metric:

- The choice of metric depends on the specific requirements of your analysis and the nature of the data
- e.g if you are more concerned about outliers, MSE or RMSE might be appropriate
- R-Squared and Adjusted are useful for understanding the proportion of variance explained by the model

### Classification

- Confusion Matrix
    - TP = True Positive - Actual Positive Predicted Positive
    - TN = True Negative - Actual Negative Predicted Negative
    - FP = False Positive - Actual Negative Predicted Positive
    - FN = False Negative - Actual Positive Predicted Negative

- Accuracy
    - Proportion of correct predictions out of all predictions.
    - Accuracy = (TP + TN) / (TP + TN + FP + FN)

- Precision
    - Out of all predicted positives, how many were actually positive.
    - Precision = TP / (TP + FP)

- Recall
    - Out of all actual positives, how many the model correctly found.
    - Recall = TP / (TP + FN)

- F1 Score
    - Balance between Precision and Recall.
    - F1 = 2 * (Precision * Recall) / (Precision + Recall)

- Specificity
    - Specificity = TN / (TN + FP)

- Negative Rate
    - Out of all actual negatives, how many were correctly predicted as negative.
    - Negative Rate = TN / (FN+TN)
