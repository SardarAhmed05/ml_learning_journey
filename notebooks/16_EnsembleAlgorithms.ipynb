### Ensemble Methods

- Ensemble method is a machine learning technique that combines several base models in order to produce one optimal predictive model.

- Steps:
    - Data
    - Learning Algorithms
    - Predictions
    - Aggregate Results

#### Types of Ensemble Algorithms
- Bagging (Bootstrap Aggregating)
- Boosting
- Stacking (Stacked Generalization)

#### Bagging
- Reduces Variance and helps avoid overfitting
- Involves creating multiple subsets of the original dataset with replacement (bootstraping)
- Training a model with each subset
- Then Aggregating their predictions
- e.g Random Forest

#### Boosting
- Boosting focuses on turning weak learners into strong ones iteratively
- Train a weak learner, find errors, train the next weak learner giving importance to those errors, Repeat.
- e.g If an observation was classified incorrectly, it tries to increase the weight of this observation and vice versa
- Examples, AdaBoost, GradientBoosting, xgBoosting



- Sequential Tree Growing with weighted samples - Boosting
    - Goes through one iteration, notices errors
    - gives more importance to the noticed errors (weak learners) in next iteration
    - Repeats

#### Stacking (Stacked Generalization)
- Stacking involves training a new model to combine the predictions of several base models.
- Involves two levels of models - base-level models and a meta-model that makes the final prediction based on the base-level models' outputs.

#### Why to use?
- Better Accuracy - Provide higher accuracy than single models
- Stability - Less prone to errors due to the variance in data
- Reduces Overfitting - Help in reducing overfitting

#### Applications
- In finance for credit scoring etc
- In healthcare for disease prediction etc
- In e-commerce for recommendation systems etc

#### Challenges and Considerations
- Complexity
- Interpretability
- Parameter Tuning
