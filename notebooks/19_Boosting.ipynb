## Boosting

- Boosting is an ensemble modeling technique that attempts to build a strong classifier from the number of weak classifiers. It is done by building a model by using weak models in series.

- e.g first tree is trained, then the second tree fits and focuses more on the weak features of the first one, and then so on.

- A supervised machine learning algorithm that combines the predictions of multiple weak models to generate a powerful ensemble model

- Boosting, as opposed to classic ensemble approaches like bagging or averaging, focuses on successively training the basic models in a way that emphasizes misclassified samples from prior iterations.

- The goal is to prioritize samples that were incorrectly categorized in previous iterations, improving model's performance iteratively

#### How does it work?
- Firstly, a model is built from the training data
- Then the second model is built which tries to correct the errors present in the first model
- This is continued and models are added until either the complete training data set is predicted correctly or the maximum number of models are added.

#### Steps

- At the start of the process, each training example is given an equal weight.
- The weighted training data is used to train a weak learner.
- The error of the weak learner on the training data is computed.
- Weights are updated according to the mistake rate of the training samples. Misclassified examples are given higher weights, whereas correctly classified examples ar given lower weights.
- Steps 2-4 are repeated several times. A new weak learner is trained on the updated weights of the training examples in each cycle
- The final model is made up of all the weak learners that were trained in the preceding steps. The accuracy of each weak learner is weighted, and the final prediction is based on the weighted total of the weak learners.


#### Advantages
- Improved performance
- Ability to handle complex data
- Robustness to Noise
- Flexibility
- Interpretability

#### Applications
- Classification problems
- Regression problems
- Natural Language processing tasks
- Image and speech recognition
- Recommendation systems
- Time Series Analysis

### Various methods for Boosting

- AdaBoost (Adaptive Boosting)

    - AdaBoost is one of the most extensively used boosting algorithms. It gives weights to each data point in the training set based on the accuracy of prior models, and then trains a new model using the updated weights. AdaBoost is very useful for classification tasks.

- Gradient Boosting:
    - Gradient Boosting works by fitting new models to the residual errors of prior models. It minimizes the loss function using gradient descent and may be applied to both regression and classification problems. Popular gradient-boosting implementations include XGBoost and LightGBM.

- CatBoost (Categorical Boosting):
    - An open-source machine learning algorithm that can handle categorical data directly and is based on gradient boosting.

- Stochastic Gradient Boosting:
    - Similar to Gradient Boosting, SGB fits each new model with random subsets of the training data and random subsets of the features. This helps to avoid overfitting and may result in improved performance.

- LPBoost (Linear Programming Boosting):
    - LPBoost is a boosting algorithm that minimizes the exponential loss function using linear programming. It is capable of handling a wide range of loss functions and may be applied to both regression and classification issues.

- TotalBoost (Total Boosting):
    - TotalBoost is an AdaBoost and LPBoost boosting method. It works by minimizing a mixture of exponential and linear programming losses, and it can increase accuracy for certain types of problems.

- xGBoost (eXtreme Gradient Boosting):
    - An optimized distributed gradient boosting library designed to be efficient, flexible, and portable.

### Boosting Algorithms vs Neural Networks.


- Can boosting algorithms be better than neural networks? It depends:
    - on the Context of the problem being solved,
    - Nature of the data,
    - Specific project requirements.

- Both have their strengths and weaknesses, and their effectiveness can vary significantly depending on the scenario.

- Boosting Algorithms:
    - Strengths:
        - Efficient with structured data
        - Handling of missing values and outliers
        - Interpretable than deep neural networks
        - Less data requirement - can perform well on small data

    - Weaknesses:
        - Scaling to large datasets - they can struggle with extremely large datsets
        - Handling of unstructured data - underperforms on unstructured data like images and text, where neural networks excel

- Neural Networks:
    - Strengths:
        - Excellence with Unstructured data
        - Scalability
        - Flexibility and adaptibility - can be adjusted for a wide variety of tasks

    - Weaknesses:
        - Data Requirements - require large amount of data
        - Interpretability - often seen as 'black boxes' due to their complexity
        - Computational Resources - computationally expensive and time consuming
