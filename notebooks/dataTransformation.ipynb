#### Data Transformation
    - Outliars skew the data, removing the outliers also transform data
    - we standardize the data with outliers, bringing the data to lower scale like 0 to 1 or -1 to 1 or any scale. Or log transform like log2 log10 etc
    - Transformation means transforming the data from its original form to a point where its easy to use - transformed form
    - We can use algorithms or transformers to transform the data. data dimensions or magnitude dont change during this

    Methods of Data Transformation
        - Standardization
            - eg StandardScaler ( -3 to 3 ), bringing the data to a specific range
        - Scaling to a known range:
            - Min Max Scaler in sklearn: if we want it to be between 0 to 1 
            - Max Absolute Scaler, [-1 to 1] or [0 to 1]
            - other scalers too
        - Kernel Sizing
        - Non Linear Transformations
        

#### Standard Scaler
    -3 to +3
#### Min Max Scalar
    -1 to 1 or 0 to 1
#### Max Absolute Scaler
    -1 to 1 and 0 to 1

These are used for Linear Data and usually called Linear Scalars

#### When we scale data, we have to check if the ML Algorithm we are supposed to use can handle negative values

    - If not, then we can use Min Max Scaler

### Standard, MinMax & MaxAbs Scaler

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler
```

```python
df = {
    'age': [25,30,35,40,45],
    'height': [165,170,175,180,185],
    'weight': [55,60,65,70,75]
}

df = pd.DataFrame(df)
df.head()
```

```python
# STANDARD SCALER

scaler = StandardScaler()

#fit the scaler on the data
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df
```

```python
# MINMAX SCALER
scaler = MinMaxScaler()

#fit the scaler on the data
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df
```

```python
## MAX ABSOLUTE SCALER
scaler = MaxAbsScaler()

#fit the scaler on the data
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df
```
