#### Data Merging

```python
# pd.merge(data1, data2, on="ID") -> Merge on basis of a specific column 
# thats in both databases e.g ID
import pandas as pd
```

```python
data1 = {
    'id': [1, 2, 3],
    'Name': ["Sardar", "Shahzad", "Shahzeb"],
    'Age': [20, 17, 10]
}
df1 = pd.DataFrame(data1)
df1
```

```python
data2 = {
    'id': [1, 2, 3],
    'City': ["Karachi", "Karachi", "Karachi"],
    'Occupation': ["ML Engineer", "Student", "Student"]
}
df2 = pd.DataFrame(data2)
df2
```

```python
merged_data = pd.merge(df1, df2, on="id", how="inner")
```

```python
merged_data
```

#### Data Concatentation

```python
import pandas as pd

data1 = {'Name': ['Ahmed', 'Fatima', 'Ali'], 'Age': [28, 32, 24]}

data2 = {'Name': ['Sara', 'Usman', 'Aisha'], 'Age': [29, 35, 27]}

data3 = {'City': ['Karachi', 'Lahore', 'Islamabad'], 'Province': ['Sindh', 'Punjab', 'Islamabad']}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)
```

```python
concatenatedRows_data = pd.concat([df1, df2])
concatenatedRows_data
```

```python
concatenated_data = pd.concat([concatenatedRows_data, df3], axis=1)
print(concatenated_data)
```

#### Drop Duplicates

```python
data = {
    "id": [1, 2, 3, 4, 5, 3, 2, 6, 7, 4, 8],
    "name": ["Alice", "Bob", "Charlie", "David", "Eva", "Charlie", "Bob", "Frank", "Grace", "David", "Hannah"],
    "marks": [85, 90, 75, 80, 95, 75, 90, 88, 92, 80, 85],
    "class": ["X", "X", "Y", "X", "Y", "Y", "X", "Z", "Y", "X", "Z"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df = df.drop_duplicates(subset="name")
```
