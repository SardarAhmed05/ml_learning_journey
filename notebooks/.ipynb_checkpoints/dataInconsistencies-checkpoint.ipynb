#### Data Inconsistencies
    1 - Inconsistent Formats ( of data e.g wrong data type or format ) 
        - converting multiple data with different formats into single non format
    2 - Inconsistent Naming Conventions
        - eg USA, US, United States, United States of America are all the same. we have to make it consistent
    3 - Typographical Errors - mistakes in data entry
        - eg Pakistan written is Paakistan or pkistan
    4 - Duplication
    5 - Contradictory Data
        - when multiple data contradicts each other e.g fathers age < sons age

```python
import pandas as pd
```

```python
data = {
    'date': ['2021-12-01', '01-12-2022', '2022/12/01', '12-01-2021'],
    'country': ['USA', 'U.S.A.', 'America', 'United States'],
    'name': ['John Doe','Jonh Doe', 'Jane Doe', 'Jane Doe'],
    'sales_2020': [100, 200, None, 200],
    'sales_2021': [None, 150, 300, 150]
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# Standardizing the date format
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df.head()
```

```python
df['date'] = df['date'].fillna(df['date'].mode()[0])
```
