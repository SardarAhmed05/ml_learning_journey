### Data Discretization
    - Data Discretization is a preprocessing technique used to transform continuous or numerical data into discrete categories or bins.

    - Categorical -> Numerical: Encoding
    - Continuous Numerical -> Discrete Categories/Bins: Discretization
    - Used when the algorithm prefers categorical data
    

```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import KBinsDiscretizer
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
# Impute missing values

df['age'] = df['age'].fillna(df['age'].median())
df['fare'] = df['fare'].fillna(df['fare'].median())
```

```python
# Age discretization 
age_discretizer = KBinsDiscretizer(n_bins=2, encode='ordinal', strategy='uniform')
df['age_bin'] = age_discretizer.fit_transform(df[['age']])
```

```python
df.head()
```

```python
df['age_bin'].value_counts()
```

```python
sns.histplot(df['age'], kde=True)
```

```python
age_discretizer.bin_edges_[0]
```

```python
sns.histplot(df, x='age', hue='age_bin')
```

```python
# Pandas method for binning
```

```python
df['age_bin2'] = pd.cut(df['age'], bins=3, labels=[0,1,2])
sns.histplot(df, x='age', hue='age_bin2')
```

```python
bins = [0,10,30,40,50,80]
labels = ['Child', 'Teenager','Young Adult','Adult', 'Elder']
df['age_bin3'] = pd.cut(df['age'], bins=bins, labels=labels)
sns.histplot(df, x='age', hue = 'age_bin3' )
```
